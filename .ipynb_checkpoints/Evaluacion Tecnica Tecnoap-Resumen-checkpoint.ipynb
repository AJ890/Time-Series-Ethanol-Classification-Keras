## Busqueda Inicial de Algoritmos de Clasificacion


```python
seed = 2
scoring = 'accuracy'
models = []

models.append(('KNN', KNeighborsClassifier(n_neighbors=5)))
models.append(('SVM', SVC(C=50, gamma= .1, kernel='rbf')))
models.append(('RandomForestClassifier', RandomForestClassifier(n_estimators=150)))
# evaluate each model in turn
results = []
names = []

for name,model in models:
    kfold = model_selection.KFold(n_splits=10, random_state=seed)
    cv_results = model_selection.cross_val_score(model,X1, y , cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg) 
```

# Comparar los 3 mejores modelos encontrados con gridsearch

```python
import numpy as np
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline

seed=1
#Create a pipeline
pipe = Pipeline([('classifier', RandomForestClassifier())])

# Create space of candidate learning algorithms and their hyperparameters
search_space = [{'classifier': [KNeighborsClassifier()],
                 'classifier__n_neighbors':[5,8,16,20]},
                {'classifier': [RandomForestClassifier(random_state=1)],
                 'classifier__n_estimators': [50,70, 100,200,300]},
                {'classifier': [SVC()],
                 'classifier__kernel': ['rbf'],
                 'classifier__C': [5,50,100,150, 500, 700, 2000],
                 'classifier__gamma': [1e-1,.015,0.05, 0.08,1e-2]}]
#Hacer grid search 
clf = GridSearchCV(pipe, search_space, cv=10, verbose=0,scoring='accuracy')

best_model = clf.fit(X1, y)

#Grabar a un archivo excel 
df_final = pd.concat([pd.DataFrame(best_model.cv_results_['rank_test_score'],columns=['Rank']),
           pd.DataFrame(best_model.cv_results_['mean_test_score'],columns=['Score']),
           pd.DataFrame(best_model.cv_results_['params'])], axis=1)


df_final.to_csv('Best_Classifier.csv')
best_model.best_params_
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Conv2D, MaxPooling2D
# more info on callbakcs: https://keras.io/callbacks/ model saver is cool too.
import tensorflow as tf
import time
import pandas as pd
from sklearn import model_selection
from sklearn.preprocessing import QuantileTransformer
from tensorflow.keras.callbacks import ModelCheckpoint
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import TensorBoard
from tensorflow.keras.callbacks import ModelCheckpoint
from sklearn import preprocessing
import numpy as np
from keras import optimizers
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
```

```python
train =  pd.read_excel('Ftrain.xlsx')
train = train.drop(train.columns[0], axis=1)
test =  pd.read_excel('FTest.xlsx')
test = test.drop(test.columns[0], axis=1)
train.head()
y1 = train[[i for  i in train.columns[27:36]]].values

#X1 = preprocessing.scale(train[[i for  i in train.columns[0:27]]])
#Preparacion de la informacion
def prepare(X11):
    qt = QuantileTransformer(n_quantiles=10, random_state=0) 
    X12= qt.fit_transform(X11) 
    X21= np.delete(X12, [11,12,20], axis=1)
    return X21

X1=prepare(train[[i for  i in train.columns[0:27]]])
```

# Definicion de Parametros de Red

```python
#tensorboard --logdir=logs/ 
first_layer=75
dense_layers = [2]
layer_sizes = [200]
conv_layers = [0]
x_train, x_test, y_train, y_test = train_test_split(X1, y1, test_size = 0.2,shuffle=False, random_state = 42)
```

```python
#Esto genera loops para realizar comparaciones para cada arquitectura
```

```python
x_train.shape[1:]
```

```python


for dense_layer in dense_layers:#loops sobre dense layers
    for layer_size in layer_sizes:#loops sobre nodos
        for conv_layer in conv_layers:
            NAME = "{}-Flnodes-{}-nodes-{}-dense-{}".format(first_layer,layer_size, dense_layer, int(time.time()))#Nombrar cada iteracion
            print(NAME)
            
            model = Sequential()
            model.add(Dense(first_layer, input_shape=x_train.shape[1:]))
            model.add(Activation('relu'))
            

            #Estas son las capas ocultas que se pueden modificar
            for _ in range(dense_layer):
                model.add(Dense(layer_size))
                model.add(Activation('relu'))
                model.add(Dropout(0.2))
                     
            
            model.add(Dense(y_train.shape[1]))
            model.add(Activation('softmax'))
            #Verifcacion en Tensorboard
            
            tensorboard = TensorBoard(log_dir="logs/{}".format(NAME))
            filepath="mod/{}".format(NAME)
            
            checkpoint = ModelCheckpoint(filepath, monitor='val_acc', verbose=1, 
                                 save_best_only=True, mode='max')
            #sgd = optimizers.SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
            #ada=optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
            
            model.compile(loss='categorical_crossentropy',
                          optimizer='adam',
                          metrics=['accuracy'])

            model.fit(x_train, y_train, epochs=50,shuffle=True, 
                      callbacks=[tensorboard,checkpoint], validation_data=(x_test, y_test)) 
```

# Por Ultimo Intentemos con Reduccion Dimensional
# Grid Search con diferentes parametros y metodos de Reduccion Dimension

```python
y1 = train[[i for  i in train.columns[27:36]]].idxmax(axis=1)
le = preprocessing.LabelEncoder().fit(y1.unique())
y = le.transform(y1)

qt = QuantileTransformer(n_quantiles=10, random_state=0)
X11=qt.fit_transform(train[[i for  i in train.columns[0:27]]])
XR=np.delete(X11, [11,12,20], axis=1)
```

```python
from __future__ import print_function, division
seed=1
pipe = Pipeline([
# populated by the param_grid
    ('reduce_dim', None),
    ('classify', SVC(kernel='rbf',gamma=0.01))
])

#Definicion de Parametros de busqueda
N_FEATURES_OPTIONS = [10,12,16,19,20,22,23]
C_OPTIONS = [20,50,350,700,2000]
param_grid = [
    {
        'reduce_dim': [PCA(iterated_power=7)],
        'reduce_dim__n_components': N_FEATURES_OPTIONS,
        'classify__C': C_OPTIONS
    },
    {
        'reduce_dim': [IncrementalPCA(batch_size=50)],
        'reduce_dim__n_components': N_FEATURES_OPTIONS,
        'classify__C': C_OPTIONS
    }
     , {
        'reduce_dim': [SelectKBest(chi2)],
        'reduce_dim__k': N_FEATURES_OPTIONS,
        'classify__C': C_OPTIONS
    }
]

reducer_labels = ['PCA', 'IncrementalPCA','Chi']

grid = GridSearchCV(pipe, cv=5, n_jobs=1, param_grid=param_grid)

grid.fit(XR, y)

COLORS = 'bgrcmyk'
mean_scores = grid.cv_results_['mean_test_score']

#loop para obtener los mejores resultados para cada C de SVM
res=[]
for i in range(len(reducer_labels)):
    re1=mean_scores.reshape(len(reducer_labels),len(C_OPTIONS)*len(N_FEATURES_OPTIONS))[i]
    res.append(re1.reshape(len(N_FEATURES_OPTIONS),len(C_OPTIONS)).max(axis=1))
    
#Visualizacion de Resultados
plt.figure(figsize=(10, 8))
for i, (label, reducer_scores) in enumerate(zip(reducer_labels, res)):
    plt.plot(np.asarray(N_FEATURES_OPTIONS),reducer_scores, label=label, color=COLORS[i])
plt.grid(zorder=1)   
plt.title("Comparacion")
plt.xlabel('Reduccion de atributos')
plt.xticks(N_FEATURES_OPTIONS)
plt.ylabel('accuracy')
plt.legend(loc='upper left')
plt.show()


#Grabar a un archivo .csv
df = pd.concat([pd.DataFrame(grid.cv_results_['rank_test_score'],columns=['Rank']),
           pd.DataFrame(grid.cv_results_['mean_test_score'],columns=['Score']),
           pd.DataFrame(grid.cv_results_['params'])], axis=1)

df.sort_values(['Rank']).to_csv('DimReduction.csv')

print('The best parameters are:',grid.best_params_)
```
